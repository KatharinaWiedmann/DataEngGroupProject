```python
ABSTAIN = -1
NOTFRESH = 0
FRESH = 1

positive_word = ['awesome','amazing','appealing','attractive','best','brilliant', 
                'bravo','cool','creative','excellent','fun','funny','good',
                'great','gorgeous','healing','impressive','innovative', 'like', 
                 'good', 'funny','fun','love', 'exciting', 'fun', 'warm', 'fantasy', 
                 'perfectly','fasinating', 'beautiful','laughs', 'humor', 'amazing', 'laugh', 'entertaining', 
                 'gleefully', 'powerful', 'friendship', 'beautiful', 'enjoyable', 'romantic', 'surprises', 
                 'clear', 'fantasy', 'entertainment', 'brilliance', 'unexpected', 'fasinating', 'definitely', 'heartwarming'
                'perfection', 'pleasure', 'brilliantly', 'hilarious', 'delightful', 'truly', 'inspiring', 'charismatic', 'humor', 
                 'superb', 'aesthetic','overwhelming', 'playful', 'chilling', 'fascination', 
                 'utterly', 'beautifully', 'brilliant', 'incredible', 'humour', 'artistically']
@labeling_function()

def good(x):
    for word in positive_word:
        if word in str(x).lower():
            return FRESH
    return ABSTAIN
#return FRESH if "best" in x.str.lower() else ABSTAIN
```

```python
negative_word = ['lack', 'lacks', 'bad', 'poor', 'poorest', 'poorer', 'disappointing', 'disappointed', 'stupid', 'nasty', 
                 'predictable', 'lacks', 'lack', 'confusing', 'bad', 'predictable', 'offensive', 'forgettable', 'lacks', 
                 'excuse', 'wrong', 'difficult', 'lost', 'empty', 'boringly', 'boring', 'badly', 'doubly', 'disappointed', 
                 'terrific', 'regrettable' ,'unattractive', 'terrible', 'awkward', 'disappointing', 'failed', 'negative', 
                 'unpleasant', 'confusing', 'violence', 'stupid', 'unfortunately', 'mistake', 'nasty', 'poorest', 'disappointment']
@labeling_function()

def bad(x):
    for word in negative_word:
        if word in str(x).lower():
            return NOTFRESH
    return ABSTAIN
```

```python
lfs = [good, bad]

applier = PandasLFApplier(lfs=lfs)
sample_L = applier.apply(sampledata)
```

```python
@preprocessor(memoize=True)
def textblob_sentiment(x):
    scores = TextBlob(str(x))
    x.polarity = scores.sentiment.polarity
    x.subjectivity = scores.sentiment.subjectivity
    return x
```

```python
@labeling_function(pre=[textblob_sentiment])
def textblob_polarity(x):
    return FRESH if x.polarity > 0.8 else ABSTAIN
```

```python
@labeling_function(pre=[textblob_sentiment])
def textblob_subjectivity(x):
    return FRESH if x.subjectivity >= 0.5 else ABSTAIN
```

```python
lfs = [textblob_polarity, textblob_subjectivity]

applier = PandasLFApplier(lfs)
L_sample = applier.apply(sampledata)
```
